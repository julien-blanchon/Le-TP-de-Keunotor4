```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
N = 1000  # taille de l’´echantillon
K = 3    # Nombre de gausienne
```

```python
π = np.random.randn(K,)     # coefficients de mixage
µ = np.random.randn(K, 2)   # moyennes des K gaussiennes
Σ = np.random.randn(K, 2)  # matrices de covariance des K gaussiennes
Σ = np.abs(Σ)
```

```python
from sklearn.datasets import make_blobs

# Generate sample data
X, y_true = make_blobs(n_samples=N, n_features=K,
                       centers=(π*µ.T).T, cluster_std=Σ, random_state=0)
X = X[:, ::-1]
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
π = np.array([1/3, 1/4, 5/12])
µ = np.array([[-1.5, 0], [0.5, 0.5], [2, 1]])
Σ = np.array(
    [
        [[1, 0.01], [0.01, 0.01]],
        [[0.1, 0.01], [0.01, 0.3]],
        [[1, 0.01], [0.01, 0.01]],
    ])
np.random.seed(94)
```

```python
def generate_data(N, pi, mu, sigma):
    K = pi.shape[0]

    X = np.zeros((N, 2))
    z = np.zeros(N, dtype=np.int32)

    for i in range(N):
        # random sample latent/unobserved variable z
        z[i] = np.random.choice(np.arange(K), p=pi)
        # random sample from z^th multivariate Gaussian
        X[i, :] = np.random.multivariate_normal(mu[z[i]], sigma[z[i]])

    return X, z
```

```python
# Generate sample data
X, y_true = generate_data(N, π, µ, Σ)



plt.scatter(X[:, 0], X[:, 1])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_true)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=K, random_state=0).fit(X)
y_predict = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_predict)
```

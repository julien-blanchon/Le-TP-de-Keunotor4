```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

np.random.seed(42)
```

```python
x = np.linspace(-1, 1, 10).reshape(-1,1)
y = -x**3 + np.random.normal(0, .1, size=(10,1))

plt.figure()
plt.scatter(x, y, marker='x')
plt.show()
```

```python
model = LinearRegression().fit(x, y)

x_line = np.linspace(-1, 1, 100).reshape(-1,1)
y_line = model.predict(x_line)

plt.figure()
plt.scatter(x, y, marker='x')
plt.plot(x_line, y_line, color='red')
plt.show()
```

```python
deg = 9

x = np.linspace(-1, 1, 10).reshape(-1,1)
X = np.concatenate([x**k for k in range(1,deg+1)], axis=1)
model = LinearRegression().fit(X, y)

print(f'Polynomial degree: {deg}')
print(f'R2 score (on training data): {model.score(X, y)}')

x_line = np.linspace(-1, 1, 100).reshape(-1,1)
X_line = np.concatenate([x_line**k for k in range(1,deg+1)], axis=1)
y_line = model.predict(X_line)

plt.figure()
plt.scatter(x, y, marker='x')
plt.plot(x_line, y_line, color='red')
plt.show()
```

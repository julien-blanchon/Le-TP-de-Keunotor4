```python
def generate_data(n_ind, n_features, sigma2_star, beta):
    X = np.random.uniform(-1, 1, size=(n_ind, n_features))
    epsilon = np.random.normal(0.0, sigma2_star, size=(n_ind,))
    y = X@beta + epsilon
    return X, y
```
